```python
import pandas as pd
data = pd.read_csv("/content/Iris.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
X = data[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
y = data[['Species']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
from six import StringIO
import pydotplus
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree

dot_data = StringIO()
filename = "speciestree.png"
featureNames = data.columns[0:5]
targetNames = data["Species"].unique().tolist()
out = tree.export_graphviz(clf,feature_names=featureNames, out_file = dot_data, class_names=np.unique(y_train), filled=True, special_characters=True, rotate=False)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png(filename)
img = mpimg.imread(filename)
plt.figure(figsize=(20,30))
plt.imshow(img, interpolation="nearest")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf = DecisionTreeClassifier(criterion="entropy", max_depth=1)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```
